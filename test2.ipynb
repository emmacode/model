```python
jaundice = r"/Users/moyin/Documents/school/project/Data/NJN/jaundice"
normal = r"/Users/moyin/Documents/school/project/Data/NJN/normal"
```

```python
os.listdir(jaundice)
```

```python
#parasitized - 1
#uninfected - 0


X = []
y = []

for filename in os.listdir(jaundice)[:1000]:
    try:
        img = cv2.imread(f"{jaundice}/{filename}")
        img = cv2.resize(img, (160,160))
        img = img/255

        X.append(img)
        y.append(0)
    except:
        pass
```

```python
len(X)
```

```python
len(y)
```

```python
#parasitized - 1
#uninfected - 0

for filename in os.listdir(normal):
    try:
        img = cv2.imread(f"{normal}/{filename}")
        img = cv2.resize(img, (160,160))
        img = img/255

        X.append(img)
        y.append(1)
    except:
        pass
```

```python
len(X), len(y)
```

```python
X_arr = np.array(X)
y_arr = np.array(y)
```

```python
X_arr.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_arr, y_arr, test_size=0.3, random_state = 9)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.callbacks import EarlyStopping
```

```python
m = mobilenet_v2.MobileNetV2(
    input_shape=(160,160,3),
    include_top=False,
    classes=2,
    classifier_activation='sigmoid'
)
m.summary()
```

```python
model = Sequential()
model.add(m)
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30, batch_size=5000, validation_split = 0.25)
```
